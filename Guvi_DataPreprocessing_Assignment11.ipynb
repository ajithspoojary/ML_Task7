**Step 1: Importing the libraries**

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
import numpy as np
```

**Step 2: Importing dataset**

```python
df = pd.read_csv("Data.csv")
df
```

```python
df.info
```

```python
df.describe()
```

**Step 3: Handling the missing data**

```python
df["Salary"] = df["Salary"].replace(to_replace=np.nan,value=df["Salary"].mean())
df["Age"] = df["Age"].replace(to_replace=np.nan,value=df["Age"].mean())
df
```

**Step 4: Encoding categorical data**

```python
label = LabelEncoder()
label.fit(df["Country"])
df["Country_en"]=label.transform(df["Country"])
df
```

**Step 5: Creating a dummy variable**

```python
purchased = pd.get_dummies(df["Purchased"],drop_first=True)
df = pd.concat([df,purchased],axis=1)
df
```

**Step 6: Splitting the datasets into training sets and Test sets**

```python
X=df[["Age","Salary"]]
Y=df["Yes"]
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.3,random_state=1)
```

**Step 7: Feature Scaling**

```python
ss = StandardScaler()
ss.fit(x_train)
ssx_train = ss.transform(x_train)
ssx_test = ss.transform(x_test)
print(ssx_train,"\n\n\n",ssx_test)
```
